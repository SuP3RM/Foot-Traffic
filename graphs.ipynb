```python
# imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Final run baseline

```python
# define data
# get data from csv file="FootTrafic_90cm_NoOnePass.csv"
df = pd.read_csv('FootTrafic data.xlsx - baseline_final_run.csv')
# print(df.head())
# print(df.describe())

# x-values are RSSI signal strength
x_pass_1 = np.float64(df['baseline_final_run_1'])
x_pass_2 = np.float64(df['baseline_final_run_2'])
x_pass_3 = np.float64(df['baseline_final_run_3'])


# if pass has nan values, ignore them
x_pass_1 = x_pass_1[~np.isnan(x_pass_1)]
x_pass_2 = x_pass_2[~np.isnan(x_pass_2)]
x_pass_3 = x_pass_3[~np.isnan(x_pass_3)]


plt.figure(1)
plt.subplot(211)
plt.plot(x_pass_1, label='baseline 1 (30s)', marker='o', markersize=2, color='red')
plt.plot(x_pass_2, label='baseline 2 (30s)', marker='x', markersize=2, color='green')
plt.plot(x_pass_3, label='baseline 3 (30s)', marker='*', markersize=2, color='blue')
plt.xlabel('# of Data Points')
plt.ylabel('RSSI (dBm)')
plt.title('RSSI Signal Strength: Room - No One')
plt.legend()
plt.grid(True)
plt.show()

# Plot a single histogram of results
fig3, ax3 = plt.subplots()
n1, bins1, pacthes3 = ax3.hist(x_pass_1, bins=20, label='pass 1 (30s)', alpha=0.5, density=True)
n2, bins2, pacthes3 = ax3.hist(x_pass_2, bins=20, label='pass 2 (30s)', alpha=0.5, density=True)
n3, bins3, pacthes3 = ax3.hist(x_pass_3, bins=20, label='pass 3 (30s)', alpha=0.5, density=True)

# plot the histogram of the RSSI values
y1 = ((1 / (np.sqrt(2 * np.pi) * np.std(x_pass_1))) *
        np.exp(-0.5 * (1 / np.std(x_pass_1) * (bins1 - np.mean(x_pass_1))) ** 2))
ax3.plot(bins1, y1, label='pass 1 fit')

y2 = ((1 / (np.sqrt(2 * np.pi) * np.std(x_pass_2))) *
        np.exp(-0.5 * (1 / np.std(x_pass_2) * (bins2 - np.mean(x_pass_2))) ** 2))
ax3.plot(bins2, y2, label='pass 2 fit')

y3 = ((1 / (np.sqrt(2 * np.pi) * np.std(x_pass_3))) *
        np.exp(-0.5 * (1 / np.std(x_pass_3) * (bins3 - np.mean(x_pass_3))) ** 2))
ax3.plot(bins3, y3, label='pass 3 fit')

plt.xlabel('RSSI (dBm)')
plt.ylabel('Frequency')
plt.title('Histogram of RSSI Signal Strength\
        \n\
        \nwith avg mean = %.2f and avg std = %.2f' % (np.mean((np.mean(x_pass_1), np.mean(x_pass_2), np.mean(x_pass_3))), np.mean((np.std(x_pass_1), np.std(x_pass_2), np.std(x_pass_3)))))
plt.legend()
plt.show()
```

'Frequency' on the y-axis represents the number of times a certain range of RSSI (Received Signal Strength Indicator) values appear in the dataset. When the density parameter `density=True` is set to True in the hist function, as in this code, the histogram is normalized to form a probability density. This means that the y-axis shows the probability density of the RSSI values, not the raw count. So, a higher value on the y-axis means that range of RSSI values is more likely to appear in the dataset, relative to ranges with lower values.
